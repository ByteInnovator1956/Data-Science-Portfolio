```python
# Step 1: EDA for MyNextMovie Recommender
# Run in a Jupyter notebook or a .py file

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from typing import List

# -------------------------
# 1. Load datasets
# -------------------------
movies = pd.read_csv("movies.csv")      # columns: movieId, title, genres
ratings = pd.read_csv("ratings.csv")    # columns: userId, movieId, rating, timestamp

# Quick peek
print("Movies:", movies.shape)
print("Ratings:", ratings.shape)
display(movies.head(3))
display(ratings.head(3))

# -------------------------
# 2. Basic info & cleaning
# -------------------------
movies.info()
ratings.info()
# If timestamp exists and you want to convert:
if 'timestamp' in ratings.columns:
    ratings['timestamp'] = pd.to_datetime(ratings['timestamp'], unit='s', errors='ignore')

# -------------------------
# 3. Unique counts
# -------------------------
n_users = ratings['userId'].nunique()
n_movies = movies['movieId'].nunique()
n_ratings = len(ratings)
print(f"Unique users: {n_users}, Unique movies: {n_movies}, Total ratings: {n_ratings}")

# -------------------------
# 4. Rating distribution
# -------------------------
print(ratings['rating'].describe())
rating_counts = ratings['rating'].value_counts().sort_index()
print(rating_counts)

plt.figure(figsize=(8,4))
sns.histplot(ratings['rating'], bins=np.arange(0.25,5.25,0.5))
plt.title("Distribution of Ratings")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()

# -------------------------
# 5. Prepare genre info
# -------------------------
# movies['genres'] are pipe-separated, e.g. "Action|Adventure|Thriller"
movies['genres'] = movies['genres'].fillna('')  # avoid NaN
movies['genre_list'] = movies['genres'].apply(lambda x: x.split('|') if x != '(no genres listed)' and x!='' else [])
# explode for per-genre analysis
movies_exploded = movies[['movieId','title','genre_list']].explode('genre_list')
movies_exploded.rename(columns={'genre_list':'genre'}, inplace=True)
movies_exploded['genre'] = movies_exploded['genre'].replace('', np.nan)
movies_exploded = movies_exploded.dropna(subset=['genre']).reset_index(drop=True)

# Unique genres
unique_genres = sorted(movies_exploded['genre'].unique())
print("Unique genres:", unique_genres)

# Movies per genre
movies_per_genre = movies_exploded['genre'].value_counts().rename_axis('genre').reset_index(name='num_movies')
display(movies_per_genre.head(10))

# -------------------------
# 6. Movie-level rating stats (avg + count)
# -------------------------
movie_stats = ratings.groupby('movieId')['rating'].agg(['mean','count']).reset_index().rename(columns={'mean':'avg_rating','count':'num_ratings'})
# Merge with movies
movies_with_stats = movies.merge(movie_stats, on='movieId', how='left')
movies_with_stats['avg_rating'] = movies_with_stats['avg_rating'].fillna(0)
movies_with_stats['num_ratings'] = movies_with_stats['num_ratings'].fillna(0).astype(int)

display(movies_with_stats.sort_values('num_ratings', ascending=False).head(10))

# -------------------------
# 7. Genre-level aggregates: avg rating & total movies
# -------------------------
# Merge exploded movies with movie_stats
genre_stats = movies_exploded.merge(movie_stats, on='movieId', how='left')
# For movies without any ratings, avg_rating will be NaN -> ignore or set 0 per decision
genre_agg = genre_stats.groupby('genre').agg(
    avg_rating = pd.NamedAgg(column='avg_rating', aggfunc='mean'),
    total_movies = pd.NamedAgg(column='movieId', aggfunc=lambda x: x.nunique()),
    total_ratings = pd.NamedAgg(column='num_ratings', aggfunc='sum')
).reset_index().sort_values('avg_rating', ascending=False)

display(genre_agg)

# -------------------------
# 8. Helper: Top N movies in a genre (popularity style)
# -------------------------
def top_n_by_genre(genre: str, min_reviews: int = 50, N: int = 10) -> pd.DataFrame:
    """
    Returns top N movies in a genre sorted by avg_rating (desc), only considering movies with >= min_reviews.
    """
    # filter exploded to the genre and join movie stats
    df = movies_exploded[movies_exploded['genre'].str.lower() == genre.lower()]
    df = df.merge(movie_stats, on='movieId', how='left')
    df = df[df['num_ratings'] >= min_reviews].sort_values(by=['avg_rating','num_ratings'], ascending=[False, False])
    # drop duplicates because a movie may appear multiple times for same genre exploded (shouldn't, but safe)
    df = df.drop_duplicates(subset=['movieId'])
    return df[['movieId','title','avg_rating','num_ratings']].head(N)

# Example usage:
print("Top Action movies (min_reviews=100, N=5):")
display(top_n_by_genre('Action', min_reviews=100, N=5))

# -------------------------
# 9. Save processed tables for later steps
# -------------------------
movies_with_stats.to_csv("movies_with_stats.csv", index=False)
movies_exploded.to_csv("movies_exploded.csv", index=False)
movie_stats.to_csv("movie_stats.csv", index=False)
```

```python
# Step 2: Popularity-Based Recommender

import pandas as pd

# We'll reuse movies_exploded and movie_stats from Step 1
# If you start fresh, load them like this:
# movies = pd.read_csv("movies.csv")
# ratings = pd.read_csv("ratings.csv")
# (then redo EDA preprocessing)

# Merge movie stats with exploded genres for recommendations
movies_exploded = pd.read_csv("movies_exploded.csv")   # from Step 1
movie_stats = pd.read_csv("movie_stats.csv")           # from Step 1

# -------------------------------
# Popularity-Based Recommendation
# -------------------------------
def popularity_recommender(genre: str, min_reviews: int = 50, N: int = 10) -> pd.DataFrame:
    """
    Returns Top-N popular movies in a given genre.
    Parameters:
        genre (str): target genre (e.g., 'Action', 'Comedy')
        min_reviews (int): minimum number of ratings to consider a movie
        N (int): number of recommendations
    """
    # Filter to the genre
    df = movies_exploded[movies_exploded['genre'].str.lower() == genre.lower()]
    
    # Join with movie stats
    df = df.merge(movie_stats, on="movieId", how="left")
    
    # Apply threshold
    df = df[df['num_ratings'] >= min_reviews]
    
    # Sort by avg_rating then num_ratings
    df = df.sort_values(by=["avg_rating", "num_ratings"], ascending=[False, False])
    
    # Keep unique movies only
    df = df.drop_duplicates(subset=["movieId"])
    
    # Final clean output
    return df[["movieId", "title", "avg_rating", "num_ratings"]].head(N).reset_index(drop=True)


# -------------------------------
# Example runs
# -------------------------------
print(" Top 5 Action Movies (min_reviews=100):")
print(popularity_recommender("Action", min_reviews=100, N=5))

print("\n Top 5 Comedy Movies (min_reviews=100):")
print(popularity_recommender("Comedy", min_reviews=100, N=5))

print("\n Top 5 Romance Movies (min_reviews=50):")
print(popularity_recommender("Romance", min_reviews=50, N=5))
```

```python
# Step 3: Content-Based Recommender (Genres similarity)

import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

# Load movies (from EDA step)
movies = pd.read_csv("movies.csv")

# ----------------------------
# 1. Prepare genre one-hot encoding
# ----------------------------
# Split genres into list
movies['genres'] = movies['genres'].fillna('(no genres listed)')
movies['genre_list'] = movies['genres'].apply(lambda x: x.split('|') if x != '(no genres listed)' else [])

# Create dummy variables (one-hot for each genre)
# Example: Action|Comedy  Action=1, Comedy=1, others=0
genres_dummies = movies['genre_list'].str.join('|').str.get_dummies()

# Combine with movies DataFrame
movies_with_genres = pd.concat([movies, genres_dummies], axis=1)

# ----------------------------
# 2. Compute cosine similarity matrix
# ----------------------------
genre_matrix = genres_dummies.values
cosine_sim = cosine_similarity(genre_matrix, genre_matrix)

# ----------------------------
# 3. Recommender function
# ----------------------------
def content_recommender(title: str, N: int = 5) -> pd.DataFrame:
    """
    Recommend Top-N movies similar in genre to the given movie title.
    """
    if title not in movies_with_genres['title'].values:
        return pd.DataFrame({"Error": [f"Movie '{title}' not found in dataset."]})
    
    # Find index of the movie
    idx = movies_with_genres.index[movies_with_genres['title'] == title][0]
    
    # Get similarity scores for this movie
    sim_scores = list(enumerate(cosine_sim[idx]))
    
    # Sort by similarity (skip self idx=0)
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    
    # Top-N (excluding itself)
    top_indices = [i for i, score in sim_scores[1:N+1]]
    
    # Return results
    return movies_with_genres.loc[top_indices, ['movieId','title','genres']].reset_index(drop=True)

# ----------------------------
# Example usage
# ----------------------------
print(" Movies similar to 'Matrix, The (1999)':")
print(content_recommender("Matrix, The (1999)", N=5))

print("\n Movies similar to 'Toy Story (1995)':")
print(content_recommender("Toy Story (1995)", N=5))

print("\n Movies similar to 'Casablanca (1942)':")
print(content_recommender("Casablanca (1942)", N=5))
```

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

# =======================
# Load Data
# =======================
ratings = pd.read_csv("ratings.csv")   # columns: userId, movieId, rating
movies = pd.read_csv("movies.csv")     # columns: movieId, title, genres

# =======================
# Step 1: Create UserMovie Rating Matrix
# =======================
user_movie_matrix = ratings.pivot_table(
    index='userId',
    columns='movieId',
    values='rating'
).fillna(0)

# =======================
# Step 2: Similarity Matrices
# =======================
# User-User Similarity
user_similarity = cosine_similarity(user_movie_matrix)
user_similarity_df = pd.DataFrame(user_similarity,
                                  index=user_movie_matrix.index,
                                  columns=user_movie_matrix.index)

# Item-Item Similarity
item_similarity = cosine_similarity(user_movie_matrix.T)
item_similarity_df = pd.DataFrame(item_similarity,
                                  index=user_movie_matrix.columns,
                                  columns=user_movie_matrix.columns)

# =======================
# Step 3: Recommend Movies (User-User CF)
# =======================
def recommend_user(user_id, num_recommendations=5):
    # Find similar users
    similar_users = user_similarity_df[user_id].sort_values(ascending=False)[1:11]
    
    # Get ratings of those users
    similar_users_ratings = user_movie_matrix.loc[similar_users.index]
    
    # Average their ratings
    avg_ratings = similar_users_ratings.mean(axis=0)
    
    # Remove movies already watched by target user
    watched_movies = user_movie_matrix.loc[user_id][user_movie_matrix.loc[user_id] > 0].index
    recommendations = avg_ratings.drop(watched_movies).sort_values(ascending=False).head(num_recommendations)
    
    return movies[movies['movieId'].isin(recommendations.index)][['movieId','title','genres']]

# =======================
# Step 4: Recommend Similar Movies (Item-Item CF)
# =======================
def recommend_item(movie_id, num_recommendations=5):
    # Find similar movies
    similar_scores = item_similarity_df[movie_id].sort_values(ascending=False)[1:num_recommendations+1]
    
    return movies[movies['movieId'].isin(similar_scores.index)][['movieId','title','genres']]

# =======================
# Example Runs
# =======================
print(" User-User CF Recommendations for User 1:")
print(recommend_user(1, 5))

print("\n Item-Item CF Recommendations for MovieId=1 (Toy Story):")
print(recommend_item(1, 5))
```

```python
#  Item-Based Collaborative Filtering (without surprise)
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

# Load dataset
movies = pd.read_csv("movies.csv")
ratings = pd.read_csv("ratings.csv")

# Create user-movie ratings matrix
user_movie_matrix = ratings.pivot_table(index="userId", columns="movieId", values="rating")

# Fill NaN with 0 for cosine similarity
movie_ratings_filled = user_movie_matrix.fillna(0)

# Compute similarity matrix between movies
similarity_matrix = cosine_similarity(movie_ratings_filled.T)
similarity_df = pd.DataFrame(similarity_matrix, 
                             index=user_movie_matrix.columns, 
                             columns=user_movie_matrix.columns)

# Function: Get top-N similar movies
def get_similar_movies(movie_title, top_n=5):
    try:
        movie_id = movies[movies['title'].str.contains(movie_title, case=False, regex=False)].iloc[0]['movieId']
    except IndexError:
        return f"Movie '{movie_title}' not found!"

    # Get similarity scores for this movie
    similar_scores = similarity_df[movie_id].sort_values(ascending=False)

    # Exclude the movie itself and get top-N
    top_similar_ids = similar_scores.drop(movie_id).head(top_n).index

    # Get movie details
    return movies[movies['movieId'].isin(top_similar_ids)][['movieId', 'title', 'genres']]

#  Example usage
print("Movies similar to 'Matrix, The (1999)':")
print(get_similar_movies("Matrix, The (1999)"))

print("\nMovies similar to 'Toy Story (1995)':")
print(get_similar_movies("Toy Story (1995)"))
```

```python
import pandas as pd

# Load datasets
movies = pd.read_csv("movies.csv")
ratings = pd.read_csv("ratings.csv")

# 1 Preprocessing: create genre list
movies['genre_list'] = movies['genres'].apply(lambda x: x.split('|'))

# 2 Calculate avg_rating and num_ratings
movie_stats = ratings.groupby('movieId')['rating'].agg(['mean','count']).reset_index()
movie_stats.columns = ['movieId','avg_rating','num_ratings']

# 3 Merge stats back to movies
movies = movies.merge(movie_stats, on='movieId', how='left')

# 4 Explode genre_list for easier filtering
movies_exploded = movies.explode('genre_list')

# 5 Function to interactively get top movies by genre
def top_movies_by_genre():
    genre = input("Enter a genre (e.g., Action, Comedy, Drama): ").strip()
    n = int(input("How many movies do you want to see? "))
    min_reviews = int(input("Minimum number of ratings a movie must have? "))

    # Filter movies by genre and min_reviews
    filtered = movies_exploded[
        (movies_exploded['genre_list'].str.lower() == genre.lower()) &
        (movies_exploded['num_ratings'] >= min_reviews)
    ]

    # Sort by avg_rating descending
    top_movies = filtered.sort_values(by='avg_rating', ascending=False).head(n)

    if top_movies.empty:
        print(f"No movies found for genre '{genre}' with min_reviews={min_reviews}")
    else:
        print(f"\nTop {n} {genre} movies (min_reviews={min_reviews}):\n")
        print(top_movies[['movieId', 'title', 'avg_rating', 'num_ratings']])

#  Call the function
top_movies_by_genre()
```
